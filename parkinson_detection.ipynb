```python
# importing dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset
data = pd.read_csv('parkinson_desease.csv')
data.head()
```

```python
# get dataset info
data.info()
```

```python
# check number of rows and columns
data.shape
```

```python
# checking the number of values in the target column
data['status'].value_counts()
```

```python
# check if there is any null in the dataset
data.isnull().sum()
```

```python
# creating a visualization of the data
data.hist(figsize=(15,15))
```

```python
# checking the data stats
data.describe()
```

```python
# dropping useless column
data = data.drop(['name'], axis=1)
```

```python
# checking the mean value for each target output
data.groupby('status').mean()
```

```python
# splitting the data in feature and target datasets
X = data.drop(['status'], axis=1)
y = data['status']
```

```python
# splitting and training datasets
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
# import module
from sklearn.svm import SVC
```

```python
# training and predicting the output
model = SVC()
model.fit(X_train, y_train)
pred = model.predict(X_train)
```

```python
# importing metrics
from sklearn.metrics import classification_report
```

```python
# applying metrics to train dataset
classification = classification_report(y_train, pred)
print(classification)
```

```python
# applying metrics to test dataset
pred_test = model.predict(X_test)
acc_test = classification_report(y_test, pred_test)
print(acc_test)
# the results are not good
```

## Implementing Standard Scaler

```python
# as the results for test dataset are not good we scale the dataset to try to improve the prediction
# importing Standard Scale module
from sklearn.preprocessing import  StandardScaler
```

```python
# scaling, training and prediction test
scale = StandardScaler()

scale.fit(X_train)

X_scaled_train = scale.transform(X_train)

X_scaled_test = scale.transform(X_test)
```

```python
# printing the scale dataset
print(X_scaled_train)
```

```python
# fitting and predicting train dataset
model = SVC()
model.fit(X_scaled_train, y_train)
pred_scaled = model.predict(X_scaled_train)
acc_test = classification_report(y_train, pred_scaled)
print(acc_test)
```

```python
# fitting and predicting test dataset
pred_scaled_test = model.predict(X_scaled_test)
acc_scale_test = classification_report(y_test, pred_scaled_test)
print(acc_scale_test)
# the standard scaler helped to improve dramatically the resultos of predictions
```
